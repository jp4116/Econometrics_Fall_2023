```python
df = pd.read_excel('PS4data.xls')
df
```

```python
df['C0'] = (df['real consumption of nondurables'] + df['real consumption of services'])/df['population']
df['C1'] = df['C0'].shift(1)
df['C2'] = df['C1'].shift(1)
df['C3'] = df['C2'].shift(1)
df['C4'] = df['C3'].shift(1)
df = df.dropna()
df
```

```python
# Linear model 
clist = df.columns[-4:].tolist()
X = df[clist]
X = sm.add_constant(X)

Y = df[df.columns[-5]]

linear_model = sm.OLS(Y,X)
model_1 = linear_model.fit()
print(model_1.summary()) 
```

```python
## Runing hypo. test
hypotheses = '(C2 = 0), (C3 = 0), (C4 = 0)'
f_test = model_1 .f_test(hypotheses)
print(f_test)
```

<br> The F test rejects the hypothesis that $\beta_2=\beta_3=\beta_4=0$.
<br> We reject the null hypothesis for the first and the fourth lags of consumption at the 1\% sig. level. On the teh other hand, we fail to reject the null hypothesis for lag 2, and 3. The significance of the first lag is consistent with Hall's hypothesis. The significance of the fourth lag is not consistent with his theory, and one can argue that it could be related to seasonality.

```python
df['C5'] = df['C4'].shift(1)
df['C6'] = df['C5'].shift(1)
df['log_c0/c1'] = np.log(df['C0']/df['C1'])
df['log_c2/c3'] = np.log(df['C2']/df['C3'])
df['log_c3/c4'] = np.log(df['C3']/df['C4'])
df['log_c4/c5'] = np.log(df['C4']/df['C5'])
df['log_c5/c6'] = np.log(df['C5']/df['C6'])
df['Y0'] = df['real disposable income']/df['population']
df['Y1'] = df['Y0'].shift(1)
df = df.dropna()
df['log_y0/y1'] = np.log(df['Y0']/df['Y1'])
df
```

```python
# First Stage
clist = ['log_c2/c3', 'log_c3/c4', 'log_c4/c5', 'log_c5/c6']
X = df[clist]
X = sm.add_constant(X)

Y = df['log_y0/y1'] 

first_stage_model = sm.OLS(Y,X)
model_fs = first_stage_model.fit(cov_type = 'HC0')
print(model_fs.summary()) 
```

```python
# Second Stage
X = model_fs.fittedvalues.tolist()
X = sm.add_constant(X)

Y = df['log_c0/c1']

second_stage_model = sm.OLS(Y,X)
model_ss = second_stage_model.fit(cov_type = 'HC0')
print(model_ss.summary()) 
```

To address the endogeneity problem, we follow the proposal of using the $\log \left(C_{t-2} / C_{t-3}\right), \log \left(C_{t-3} / C_{t-4}\right), \log \left(C_{t-4} / C_{t-5}\right)$, and $\log \left(C_{t-5} / C_{t-6}\right)$ as instruments, and run the 2SLS. 
<br> From the results, we see that the coefficient is statistically different from 0 at 5\% sig level, which means that 43.7\% of the households are keynesian (i.e. consume their current disposable income directly) instead of consuming according to the permanent income hypothesis.

We cannot reject the null hypothesis that $\beta_2=0$ (control variable in our model). Therefore, we cannot reject the hypothesis of exogeneity.

```python
#### Overidentification test

# instead of taking the residuals of the second stage, we recompute using the endogenous variable

res = df['log_c0/c1'] -  model_ss.predict(sm.add_constant(df['log_y0/y1']))

clist = ['log_c2/c3', 'log_c3/c4', 'log_c4/c5', 'log_c5/c6']
X = df[clist]
X = sm.add_constant(X)

over_ident_model = sm.OLS(res,X)
model_over = over_ident_model.fit()
print(model_over.summary()) 

214*model_over.rsquared
```

To test for over-identification, we recompute residuals using the endogenous variable (variable res in our model above) and regress it on the instruments to get the $R^2$.
<br> The Sargan's test statistic is: $S= nR^2 = 214*0.036=7.738$
<br> The asymptotic dist under $H_0$ (valid over-identifying restrictions) is $\chi^2(l-k)=\chi^2(4-1)$, that has critical value is $7.815$ at 5% of sig. level. Therefore, we cannot reject the null hypothesis of valid over-identifying restrictions. 

<br> Note that the F-stat from the first stage is $3.208 <10$, which suggests that it is a weak instrument according to Stock-Yogo. So it can be tha case that testing for endogeneity and the over-identifying restrictions may be flawed.
